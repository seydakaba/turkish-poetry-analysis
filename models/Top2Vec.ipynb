```python
import re
import csv
import pandas as pd
import numpy as np
from top2vec import Top2Vec
```

```python
file = open('../allPoetry.txt', encoding="utf8")
```

```python
allPoems=pd.read_csv("../allPoetry.txt",sep= " \t" ,error_bad_lines=False, engine = 'python',encoding = "utf-8")
```

```python
df=pd.DataFrame()
df["allpoems"]=allPoems
```

```python
docs=list(allPoems.loc[:, "allpoems"].values)
```

```python
model=Top2Vec(docs)
```

```python
topic_sizes, topic_nums = model.get_topic_sizes()
```

```python
topic_words, word_scores, topic_nums = model.get_topics()
```

```python
for topic in topic_nums:
    model.generate_topic_wordcloud(topic)
```

```python
words, word_scores = model.similar_words(keywords=["kar"], keywords_neg=[], num_words=20)
for word, score in zip(words, word_scores):
    print(f"{word} {score}")
```
